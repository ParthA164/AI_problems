```python
import copy

q=[]
visited=[]

# there are 2 jugs capacity of 1st jug is 4 and that of second is 3
# we want 2 litres of water in first jug 
def compare(s):
    if s[0]==2:
        return 1
    else:
        return 0      

            
def full(s):
    s1=copy.deepcopy(s)
    s1[0]=4
    s1[1]=3
    return s1

def empty(s):
    s1=copy.deepcopy(s)
    s1[0]=0
    s1[1]=0
    return s1

def fill_4(s):
    s1=copy.deepcopy(s)
    s1[0]=4
    return s1

def fill_3(s):
    s1=copy.deepcopy(s)
    s1[1]=3
    return s1

def empty_4(s):
    s1=copy.deepcopy(s)
    s1[0]=0
    return s1

def empty_3(s):
    s1=copy.deepcopy(s)
    s1[1]=0
    return s1

def transfer_to_4(s):
    s1=copy.deepcopy(s)
    
    if s1[0]==4:
        return s1
    else:
        sum=s1[0]+s1[1]
        if sum>=4:
            s1[0]=4
            s1[1]=sum-4
        else:
            s1[0]=sum
            s1[1]=0
    
    return s1

def transfer_to_3(s):
    s1=copy.deepcopy(s)
    
    if s1[1]==3:
        return s1
    else:
        sum=s1[0]+s1[1]
        if sum>=3:
            s1[1]=3
            s1[0]=sum-3
        else:
            s1[1]=sum
            s1[0]=0
    
    return s1

def generate_child(s):
    global q
    global visited

    new_state=full(s)
    if new_state not in visited and new_state not in q:
        q.append(new_state)

    new_state=empty(s)
    if new_state not in visited and new_state not in q:
        q.append(new_state)

    new_state=fill_4(s)
    if new_state not in visited and new_state not in q:
        q.append(new_state)

    new_state=fill_3(s)
    if new_state not in visited and new_state not in q:
        q.append(new_state)

    new_state=transfer_to_4(s)
    if new_state not in visited and new_state not in q:
        q.append(new_state)

    new_state=transfer_to_3(s)
    if new_state not in visited and new_state not in q:
        q.append(new_state)

    new_state=empty_3(s)
    if new_state not in visited and new_state not in q:
        q.append(new_state)

    new_state=empty_4(s)
    if new_state not in visited and new_state not in q:
        q.append(new_state)


def search():
    global q
    global visited
  
    while q:
        
        current_state=q[0]
        del q[0]

        if compare(current_state):
            visited.append(current_state)
            print("Found\n")
            return
        else:
            visited.append(current_state)
            generate_child(current_state)
            


def main():
    global q
    global visited

    s=[0,0] #s[0] represents water in 4 litre jug and s[1] represents water in 3 litre jug.
    
    q.append(s)

    search()

    l=len(visited)
    print(visited)
    print('\n');
    print("The number of states it visited is: ",l)
    
if __name__=="__main__":
    main()

```
